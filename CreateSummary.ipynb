```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv('OPENAI_API_KEY')
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", max_tokens=1000, temperature=0)
```

```python
template = """
You are a helpful AI assistant. Your task is to sumarize a podcast transcript in 3 paragraphers at the most.

Transcript: {transcript}

Please provide a clear and concise answer:
"""

prompt = PromptTemplate(template=template, input_variables=["transcript"])
```

```python
templateTopics = """
Generate the 5 most interasting topics relative to Artificial Intelligence.

Transcript: {transcript}

Please provide your response only as a json format that list relevant topics.

"""

promptTopics = PromptTemplate(template=templateTopics, input_variables=["transcript"])
```

```python
qa_chain = prompt | llm 
chainTopics = promptTopics | llm
```

```python
def get_answer(transcripts):
    input_variables = {"transcript": transcripts}
    response = qa_chain.invoke(input_variables).content
    responseTopics = chainTopics.invoke(input_variables).content
    return response, responseTopics
```
